# Creating **List**

## Python Lists

Python lists are ordered, changeable (mutable), and allow duplicate members. They are used to store multiple items in a single variable.

**Key Characteristics:**
*   **Ordered:** The items have a defined order, and that order will not change. If you add new items to a list, the new items will be placed at the end of the list.
*   **Changeable (Mutable):** You can change, add, and remove items after the list has been created.
*   **Allow Duplicates:** Since lists are indexed, they can have items with the same value.
*   **Indexed:** List items are indexed, and you can access them by referring to the index number.

**Common List Operations:**
*   **Creating a List:**
    ```python
    my_list = [1, 'hello', 3.14, True]
    empty_list = []
    ```
*   **Accessing Elements:** (Using indexing)
    ```python
    my_list = [10, 20, 30, 40]
    print(my_list[0]) # Output: 10
    print(my_list[1]) # Output: 20
    print(my_list[-1]) # Output: 40 (last element)
    ```
*   **Slicing Lists:**
    ```python
    my_list = [10, 20, 30, 40, 50]
    print(my_list[1:4]) # Output: [20, 30, 40]
    print(my_list[:3])  # Output: [10, 20, 30]
    print(my_list[2:])  # Output: [30, 40, 50]
    ```
*   **Adding Elements:**
    *   `append()`: Adds an element to the end of the list.
        ```python
        my_list = [1, 2]
        my_list.append(3) # my_list is now [1, 2, 3]
        ```
    *   `insert()`: Adds an element at a specified index.
        ```python
        my_list = [1, 3]
        my_list.insert(1, 2) # my_list is now [1, 2, 3]
        ```
    *   `extend()`: Adds the elements of another iterable (like another list) to the end of the current list.
        ```python
        list1 = [1, 2]
        list2 = [3, 4]
        list1.extend(list2) # list1 is now [1, 2, 3, 4]
        ```
*   **Removing Elements:**
    *   `remove()`: Removes the first occurrence of a specified value.
        ```python
        my_list = [1, 2, 3, 2]
        my_list.remove(2) # my_list is now [1, 3, 2]
        ```
    *   `pop()`: Removes the item at a specified index (or the last item if no index is specified) and returns it.
        ```python
        my_list = [1, 2, 3]
        removed_item = my_list.pop(1) # removed_item is 2, my_list is [1, 3]
        ```
    *   `del` keyword: Removes item at a specified index or deletes the entire list.
        ```python
        my_list = [1, 2, 3]
        del my_list[0] # my_list is now [2, 3]
        # del my_list # Deletes the entire list
        ```
    *   `clear()`: Empties the list.
        ```python
        my_list = [1, 2, 3]
        my_list.clear() # my_list is now []
        ```
*   **Concatenating Lists:** (Using `+` operator)
    ```python
    list1 = [1, 2]
    list2 = [3, 4]
    combined_list = list1 + list2 # combined_list is [1, 2, 3, 4]
    ```
*   **List Length:**
    ```python
    my_list = [1, 2, 3]
    print(len(my_list)) # Output: 3
    ```

At the start of the day you have a checklist The tasks which you were able to finish, should get added to completed_tasks The tasks which you were not able to finish, should get added to incomplete_tasks This project is about organizing your daily tasks into two categories: completed tasks and incomplete tasks. Here’s how it works: 1. At the start of the day, you create a checklist of tasks you want to accomplish. 2. At the end of the day, you review which tasks you could finish and which you couldn’t. – If a task is finished, you move it to the completed tasks list. – If a task is not finished, you move it to the incomplete tasks list. This way, by the end of each day, you’ll have a clear overview of what you completed and what still needs attention.

````python
daily_tasks = ["data collection", "data cleaning", "data storing", "data analysis", "report generation", "feedback", "corrections"]
todo_list = [1,1,0,1,1,0,0]
completed_tasks = []
incomplete_tasks = []
for i in range(len(daily_tasks)):
  if todo_list[i]==1:
    completed_tasks.append(daily_tasks[i])
  else:
    incomplete_tasks.append(daily_tasks[i])
print ("completed List = ", completed_tasks)
print ("Incomplete list = ", incomplete_tasks)


````
