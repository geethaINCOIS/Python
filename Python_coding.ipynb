### **Functions**

```
class Solution(object):
    def buddyStrings(self, s, goal):
        """
        :type s: str
        :type goal: str
        :rtype: bool
        """

        # Length check
        if len(s) != len(goal):
            print("length of 2 strings are not equal")
            return False

        # Constraint check
        if not (1 < len(s) <= 2 * 10**4):
            return False

        # Lowercase check
        if not (s.islower() and goal.islower()):
            print("not in lowercase")
            return False


        diff = []
        for i in range(len(s)):
            if s[i] != goal[i]:
                diff.append(i)
                if len(diff) > 2:
                    return False

        if len(diff) == 2:
            i, j = diff
            return s[i] == goal[j] and s[j] == goal[i]

        # Case 2: strings already equal
        elif len(diff) == 0:
            seen = set()
            for ch in s:
                if ch in seen:
                    return True
                seen.add(ch)
            return False
        else :
            return False
# Object creation and function call
obj = Solution()
print(obj.buddyStrings("ac", "ab"))
print(obj.buddyStrings("ab", "ab"))
print(obj.buddyStrings("ba", "ab"))
print(obj.buddyStrings("aa", "aa"))
```
